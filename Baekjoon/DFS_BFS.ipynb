```python
#1260
from collections import deque

node, edge, start = map(int, input().split()) #정점의 개수, 간선의 개수, 시작 정점
graph = [[] for _ in range(node + 1)] #그래프
visited_dfs = [False] * (node + 1) #DFS에 사용할 방문정보
visited_bfs = [False] * (node + 1) #BFS에 사용할 방문정보

#노드의 연결 정보를 리스트에 저장
for i in range(edge):
    node1, node2 = map(int, input().split())
    graph[node1].append(node2)
    graph[node2].append(node1)

#방문할 정점이 여러 개인 경우 정점 번호가 작은 것부터 방문하기 위해 정렬
for g in graph:
    g.sort()

#DFS
def dfs(graph, v, visited):
    #현재 노드를 방문처리
    visited_dfs[v] = True
    print(v, end = ' ')
    
    #방문하지 않은 노드에 대해, 현재 노드와 연결된 다른 노드를 재귀적으로 방문
    for i in graph[v]:
        if not visited_dfs[i]:
            dfs(graph, i, visited)

#BFS
def bfs(graph, v, visited):
    queue = deque([v])
    #현재 노드를 방문처리
    visited_bfs[v] = True
    
    #큐가 빌 때까지 반복
    while queue:
        #큐에서 하나의 원소를 뽑아 출력
        v = queue.popleft()
        print(v, end = ' ')
        
        #방문하지 않은 노드에 대해, 해당 원소와 연결된 원소들을 큐에 삽입
        for i in graph[v]:
            if not visited_bfs[i]:
                queue.append(i)
                visited_bfs[i] = True

print("--------------")
dfs(graph, start, visited_dfs)
print("")
bfs(graph, start, visited_bfs)
```

```python
#2606
from collections import deque

node = int(input()) #컴퓨터의 수
edge = int(input()) #직접 연결된 컴퓨터 쌍의 수
graph = [[] for _ in range(node + 1)] #그래프
visited = [False] * (node + 1) #방문 정보

#노드의 연결 정보를 리스트에 저장
for i in range(edge):
    node1, node2 = map(int, input().split())
    graph[node1].append(node2)
    graph[node2].append(node1)

#BFS
def bfs(graph, v, visited):
    queue = deque([v])
    visited[v] = True
    count = 0
    
    while queue:
        v = queue.popleft()
        count += 1
        
        for i in graph[v]:
            if not visited[i]:
                queue.append(i)
                visited[i] = True
    print("--------------")
    print(count - 1) #1번 컴퓨터는 제외

bfs(graph, 1, visited)
```

```python
#2667
scale = int(input()) #지도의 크기
graph = [] #지도
count = 0 #단지내 집의 수
answer = [] #각 단지내 집의 수들
result = 0 #총 단지 수

for i in range(scale):
    graph.append(list(map(int, input())))

def dfs(x, y):
    global count
    if x <= -1 or x >= scale or y <= -1 or y >= scale:
        return False
    
    if graph[x][y] == 1:
        graph[x][y] = 0
        
        dfs(x - 1, y)
        dfs(x, y - 1)
        dfs(x + 1, y)
        dfs(x, y + 1)
        
        count += 1
        return True
    return False

for i in range(scale):
    for j in range(scale):
        if dfs(i, j) == True:
            result += 1 #단지 수 추가
            answer.append(count) #단지 내 집의 수 추가
            count = 0

print("--------------")
print(result)
answer.sort()
for i in answer:
    print(i)
```
